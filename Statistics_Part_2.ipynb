#1- Hypothesis Testing in Statistics
Hypothesis testing is a statistical method used to make inferences or decisions about a population based on sample data. It helps determine whether there is enough evidence to reject a given assumption (the null hypothesis).

#2- Null Hypothesis vs. Alternative Hypothesis
Null Hypothesis (ùêª0H 0 ): A statement that assumes no effect, difference, or relationship exists (e.g., "The average test score is 70").
Alternative Hypothesis (ùêªùëéH a ): The hypothesis that contradicts ùêª0H
and suggests an effect or difference (e.g., "The average test score is not 70").
üîπ Key Difference: The null hypothesis assumes the status quo, while the alternative hypothesis suggests a change or effect.

#3- Significance Level in Hypothesis Testing (Œ±)
The significance level (ùõºŒ±) is the probability of rejecting the null hypothesis when it is actually true (Type 1 error).
Common values are:
ùõº=0.05Œ±=0.05 (5% chance of wrongly rejecting ùêª0)ùõº=0.01
(1% chance, more strict)
üîπ Importance: It defines the threshold for statistical significance.

#4- P-Value in Hypothesis Testing
The P-value measures the probability of observing results as extreme as the sample data, assuming ùêª0H 0 is true.

üîπ Interpretation:

If P-value < ùõº Œ±: Reject ùêª0H 0  (strong evidence against ùêª0H 0 )
If P-value Œ±: Fail to reject ùêª0(insufficient evidence)

#5- Type 1 and Type 2 Errors
Type 1 Error (Œ±): Rejecting ùêª0
when it is true (false positive).
Type 2 Error (Œ≤): Failing to reject H0
when it is false (false negative).
üîπ Trade-off: Lowering Type 1 error increases the risk of Type 2 error and vice versa.

#6- One-Tailed vs. Two-Tailed Tests
One-Tailed Test: Tests for an effect in one direction (e.g., "Is the average score greater than 70?").
Two-Tailed Test: Tests for an effect in both directions (e.g., "Is the average score different from 70?").
üîπ Key Difference: One-tailed tests focus on a specific direction, while two-tailed tests consider both possibilities.

#7 Z-Test in Hypothesis Testing
A Z-test is used when:

The sample size is large (ùëõ‚â•30).

üîπ Application: Comparing sample means to population means.

#9  T-Distribution vs. Normal Distribution
The T-distribution is used when the sample size is small (n<30) and population standard deviation (œÉ) is unknown.
It is wider than the normal distribution, accounting for more
 uncertainty in small samples.

#11 T-Test in Hypothesis Testing
Used when comparing sample means:

One-Sample T-Test: Compares sample mean to a known population mean.
Two-Sample T-Test: Compares means of two independent samples.
Paired T-Test: Compares means of paired (dependent) samples.
üîπ Example: Comparing test scores of students before and after a training program.

#12 Relationship Between Z-Test and T-Test
Both are used for hypothesis testing of means:

Z-test is used for large samples with known standard deviation.
T-test is used for small samples or when standard deviation is unknown.

#13  Confidence Interval (CI) and Its Interpretation
A confidence interval provides a range of values where the true population parameter likely lies.

95% CI means "We are 95% confident the true mean is in this range."

```python
1
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))  # Two-tailed test

    print(f"Z-Score: {z_score:.3f}")
    print(f"P-Value: {p_value:.3f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0). The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis (H0). No significant difference detected.")

# Simulating random sample data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)
one_sample_z_test(sample_data, pop_mean=50, pop_std=10)
```

```python
3
import matplotlib.pyplot as plt

# Define critical z-values for a two-tailed test
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)  # Two-tailed test

# Plot decision regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label="Standard Normal Distribution")
plt.axvline(-z_critical, color='red', linestyle='dashed', label=f"Critical Value (-{z_critical:.2f})")
plt.axvline(z_critical, color='red', linestyle='dashed', label=f"Critical Value ({z_critical:.2f})")
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3)

plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.title("Two-Tailed Z-Test Decision Regions")
plt.legend()
plt.show()
```

```python
4
def plot_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.plot(x, y, label="Null Distribution (H0)")
    plt.fill_between(x, y, where=(x > norm.ppf(1-alpha)), color='red', alpha=0.5, label="Type 1 Error (Œ±)")

    alt_x = np.linspace(-4, 4, 1000)
    alt_y = norm.pdf(alt_x, loc=1.5)  # Shifted for alternative hypothesis

    plt.plot(alt_x, alt_y, label="Alternative Distribution (H1)")
    plt.fill_between(alt_x, alt_y, where=(alt_x < norm.ppf(beta, loc=1.5)), color='blue', alpha=0.5, label="Type 2 Error (Œ≤)")

    plt.xlabel("Test Statistic")
    plt.ylabel("Probability Density")
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.legend()
    plt.show()

plot_errors()
```

```python
5
from scipy.stats import ttest_ind

# Generate random data for two independent groups
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 12, 30)

t_stat, p_value = ttest_ind(group1, group2)
print(f"T-Statistic: {t_stat:.3f}, P-Value: {p_value:.3f}")

if p_value < 0.05:
    print("Reject the null hypothesis. The two groups have significantly different means.")
else:
    print("Fail to reject the null hypothesis. No significant difference detected.")
```

```python
8
def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error
    margin_error = norm.ppf((1 + confidence) / 2) * std_err  # Z-score for 95%

    ci_lower = mean - margin_error
    ci_upper = mean + margin_error
    print(f"{confidence*100}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# Example usage
sample = np.random.normal(50, 10, 30)
confidence_interval(sample)
```

```python
13
def z_scores(data):
    mean, std = np.mean(data), np.std(data, ddof=1)
    z = (data - mean) / std
    plt.hist(z, bins=20, color='blue', alpha=0.6)
    plt.title("Z-Score Distribution")
    plt.xlabel("Z-Score")
    plt.ylabel("Frequency")
    plt.show()
    return z

data = np.random.normal(100, 15, 50)
z_scores(data)
```

```python
15
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    sem = stats.sem(sample)
    interval = stats.norm.interval(confidence, loc=mean, scale=sem)
    return interval

sample_data = np.random.normal(100, 15, 30)
ci = confidence_interval(sample_data)
print(f"Confidence Interval: {ci}")
```
